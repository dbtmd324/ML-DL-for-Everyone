```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# basic lists to start
x_data = [1, 2, 3] # hours for studying
y_data = [2, 4, 6] # Grades
w_list = [] # x-axis
mse_list = [] # y-axis

# our model for the forward pass
def forward(x):
    return x * w

# loss function
def loss(x, y):
    y_pred =forward(x)
    return (y_pred - y) * (y_pred - y)

# compute loss for w
for w in np.arange(0.0, 4.1, 0.1):
    print("w=", w)
    l_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l 
        print("\t", x_val, y_val, y_pred_val, l)

    print("MSE=", l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3)

# plot the graph
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('Weight')
plt.show()
```
